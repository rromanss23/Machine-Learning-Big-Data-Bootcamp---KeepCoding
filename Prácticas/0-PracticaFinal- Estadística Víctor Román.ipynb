### Cual es el precio medio de alquiler de cada uno, la diferencia que hay es estadsticamente significativa?
Con que test lo comprobaras?

```r
aggregate(airbnb$Precio ~ airbnb$TipoAlquiler ,data=airbnb, mean)
```

```r
ggplot(data=airbnb, aes(x=Precio, colour=TipoAlquiler)) + geom_density(bw=20)
```

### Respuesta:

Como podemos ver en las grficas, las distribuciones estan muy sesgadas, por lo tanto, la media no es un estimador estadistico significativo. Para estos casos, la mediana es mas relevante estadisticamente hablando. Por lo tanto tambin la calcularemos

```r
aggregate(airbnb$Precio ~ airbnb$TipoAlquiler ,data=airbnb, median)
```

Comprobaremos si la diferencia de medias es estdisticamente significativa realizando un test de normalidad. 

```r
shapiro.test(airbnb$Precio[airbnb$TipoAlquiler == 'Entire home/apt'][1:5000])
```

```r
shapiro.test(airbnb$Precio[airbnb$TipoAlquiler == 'Private room'][1:5000])
```

```r
shapiro.test(airbnb$Precio[airbnb$TipoAlquiler == 'Shared room'])
```

### Respuesta: 
Al salir los valores muy inferiores a 0.05, podemos asegurar que los precios dependiendo del tipo de alquiler no siguen una distribucion normal, como ya se haba intuido de la representacin de la distribucin de sus valores. 

Para calcular de manera explicita si la diferencia de las medias es estadsticamente relevante, utilizaremos el test de Kruskal. El test de Kruskal es una variante del test ANOVA, indicada para comparar las medias de mas de dos grupos de datos cuyos valores no diguen una distribucion normal.

Al ser el p-valor tan sumamente bajo (p-value < 2.2e-16) podemos rechazar la hipotesis nula y por tanto aseguramos los grupos poseen diferente media y por tanto su diferencia no es estadsticamente relevante.

```r
kruskal.test( Precio ~TipoAlquiler, data=airbnb)
```

### Cuales son los 5 barrios que tienen un mayor precio medio de alquiler para apartamentos enteros?

Cual es su precio medio?

Ayuda: Usa la funcin aggregate `aggregate(.~colname,df,mean,na.rm=TRUE)`

```r
price_mean <- aggregate(airbnb_entire$Precio ~ airbnb_entire$Barrio ,data=airbnb_entire, mean, na.rm=TRUE)

newnames<-c("Barrio","Precio")

names(price_mean) = newnames
```

### Cuantos apartamentos hay en cada uno de esos barrios?

Mostrar una dataframe con el nombre del barrio, el precio y el nmero de entradas.

Ayuda: Podeis crear un nuevo dataframe con las columnas "Barrio" y "Freq" que contenga el nmero de entradas en cada barrio y hacer un merge con el dataframe del punto anterior.

```r
freq <- data.frame(table(airbnb_entire$Barrio))
names(freq) <- c('Barrio', 'Freq')
head(freq)
```

```r
mean_price_apt <- merge(x = final_price_mean, y = freq, by = "Barrio", all.x = FALSE)
mean_price_apt <- mean_price_apt[order(mean_price_apt$Precio, decreasing=TRUE),]
mean_price_apt[1:5,]
```

### Cual es la probabilidad de encontrar, en el barrio de Sol, un apartamento en alquiler con 3 dormitorios?
Cual es el margen de error de esa probabilidad?

```r
n <- nrow(barrio_sol_vbles)
ns <- nrow(barrio_sol_vbles[barrio_sol_vbles$NumDormitorios ==3, ])
prob <- ns/n
print(paste("Probabilidad:",prob," xitos:",ns," Eventos totales:",n))
```

#### Respuesta: 

Se trata experimento de una distribucin binomial, es decir, una generalizacion de Bernouilli para n sucesos independientes, cada unos con posibles resultados: Si  No, es un apartamento de 3 dormitorios.

Por tanto, el intervalo de confianza no es simtrico, y lo calcularemos mediante la funcin binom.test.

Obtenemos unos resultados de probabilidad de xtio del 7.845%, con un intervalo de confianza del 95% [0.0596,0.1]

```r
binom.test(ns,n)
```
